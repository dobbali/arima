```python
data_arima_fb_series1 = pd.Series(data_arima_fb_series['impression_cnt'].values, index=data_arima_fb_series['WkTY'])
```

```python
autocorrelation_plot(data_arima_fb_series1)
plt.show()
```

```python
data_arima_fb_series1 = data_arima_fb_series1.astype("float")
```

```python
# fit model
model = ARIMA(data_arima_fb_series1, order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
pyplot.show()
residuals.plot(kind='kde')
pyplot.show()
print(residuals.describe())
```

```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
 
X = data_arima_fb_series1.values
size = int(len(X) * 0.60)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print(data_arima.iloc[size + t])
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
